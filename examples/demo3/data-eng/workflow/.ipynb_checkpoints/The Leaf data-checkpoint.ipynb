
## Correlation matrix


Let's see if the numerical variables are correlated using Pearson metric (for numerical features).

The usefullness of finding out correlated features and remove them have some benefits:

- Avoid the curse of dimensionality: less features lead to faster training and less complex models.

- Avoid multicollinearity:  For most linear models, having correlated features leads to overfitting and numerical inacurrate results due to redundancy.

As Occam'razer stated: among several solutions for a problem, we should pick the simpler one.

On the other hand, there are some scenarios when removing correlated features is not good.

- For interpretation: the more variables, the better explanation the model can yield.

- Bayes-based models: Due to the nature of bayes theorem, that one random variable can depend of previous variables leads to dependency among features, so that having correlated variables is fine.

- Regression problems: In regression models we need highly correlated features with the target. Removing one of them can decrease the performance drastically.

For instance, we can obseve that **average_contrast** and **average_intensity** are highly correlated, so likely to be removed.

```python
%%time
from collections import OrderedDict

class_name = 'species'
features = leaf.loc[:, leaf.columns != class_name]
dictFI = OrderedDict(features)

features = list(dictFI.keys())
features.append('species')

def plot_heatmap(df):
    fig, axes = plt.subplots(figsize=(10,10))
    _ = sns.heatmap(df, annot=True, cmap="YlGnBu")

    
plot_heatmap(leaf[features].corr(method='pearson'))
```


## Feature Importance

Bagged decision trees such as Random Forest and Extra Trees can be used to estimate the importance of features. As well as, the Gradient Boosting algorithm.


```python
from sklearn.model_selection import cross_val_score

def modelfit(alg, dtrain, predictors, performCV=True, printFeatureImportance=True, cv_folds=5):
    alg.fit(dtrain[predictors], dtrain['species'])
        
    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]
    
    if printFeatureImportance:
        feat_imp = pd.Series(alg.feature_importances_, predictors).sort_values(ascending=False)
        feat_imp.plot(kind='bar', title='Feature Importances', figsize=(14.,7.))
        plt.ylabel('Feature Importance Score')
        plt.show()
        plt.close()
    
    return feat_imp
```


### ExtraTreesClassifier


```python
%%time
from sklearn.ensemble import ExtraTreesClassifier
target = ['species', 'specimen_number']


predictors = [x for x in leaf if x not in target]
model = ExtraTreesClassifier(n_estimators=20, random_state=10)
fi = modelfit(model, leaf, predictors)
print(fi.head(6))
```


## GradientBoostingClassifier


```python
%%time
from sklearn.ensemble import GradientBoostingClassifier
target = ['species', 'specimen_number']


predictors = [x for x in leaf if x not in target]
model = GradientBoostingClassifier(random_state=10)
fi = modelfit(model, leaf, predictors)
print(fi.head(6))

```


## Split into Train and Test Sets


```python
%%time

random_state = 42
np.random.seed(random_state)

from sklearn.linear_model import LogisticRegression
from sklearn import model_selection

test_size = 0.20

# RANDOM_SEED = 42
# random_state=np.random.RandomState(RANDOM_SEED)

X = X.astype(float)
# X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y,
#                                                                     test_size=test_size, random_state=random_state)
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y,
                                                                    test_size=test_size)
```

# Modeling using Pipelines

The Pipeline is a way of bringing together transformers (eg, Scaler, SelectKbest) and estimators (ExtraTrees, Logistic Regression) in a single flow, so that each link in the pipeline runs sequentially until the predictive model is achieved. For example, Transformers: Scale, Normalization, etc, and in another pipeline, transformers for features selections: univariate, pca, etc. can be put into a flow. and in another pipeline put estimators: PCA, k-means, etc. Then with the FeatureUnion function, we can join those three pipelines to create a total pipeline, which performs all those steps as one. This helps prevent data loss and increase readability.


```python
import time

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier


transformers = []
# transformers.append(( 'standardize' , StandardScaler()))
# # estimators.append(( 'normalizer' , Normalizer()))
# estimators.append(( 'lda' , LinearDiscriminantAnalysis()))
# transf_union = FeatureUnion(transformers)
# model = Pipeline(estimators)

# create feature union
selectors = []
# features.append(('transf_union', transf_union))
# pca = PCA(n_components=9, random_state=random_state)
pca = PCA(n_components=9)
selectors.append(( 'pca', pca))
# selectors.append(( 'select_best' , SelectKBest(k=3)))
# features_union = FeatureUnion(feature_eng)
# create pipeline
estimators = []
# estimators.append(( 'Features_union' , features_union))
# et = ExtraTreesClassifier(n_estimators=20, random_state=random_state)
et = ExtraTreesClassifier(n_estimators=20, n_jobs=1)
estimators.append(( 'ET' , et))
# estimators.append(('svm' , SVC(kernel="linear")))
model = Pipeline([ ('selectors', FeatureUnion(selectors)),
                  ('estimators', et)])
```

## Evaluation on training data, Cross validation (to test out different models)

```python
scoring = 'accuracy'

num_folds = 10

kfold = KFold(n_splits=num_folds, random_state=random_state)

start = time.time()
        
params = dict()

for k, v in model.get_params().items():
    # params[name+'__'+k] = [v]
    params[k] = [v]
    
grid_search_t = GridSearchCV(model, params, n_jobs=-1,
                 verbose=1, cv=kfold, return_train_score=True,
                             scoring=scoring)
grid_search_t.fit(X_train, y_train)

end = time.time()
duration = end - start

mean = grid_search_t.cv_results_['mean_test_score'][0]
std = grid_search_t.cv_results_['std_test_score'][0]

print(f'Mean accuracy: {mean} | Std accuracy: {std}')
```

## Evaluation on test dataset (real evaluation)

```python
model.fit(X_train, y_train)
result = model.score(X_test, y_test.values)
print('Score:', result)
predictions = model.predict(X_test)
```

```python
print(f'test: {y_test[:10].values}')
print(f'predictions: {predictions[:10]}')
```


### Confusion Matrix


# Conclusions


1. We must first ask ourselves, what problem do we want to solve.
2. Then plot the raw data and see relationships and patterns, to determine
If the problem can be treated as classification or regression.
3. Once we have clear relationships between the variables, we pass
to clean the data of noise, anomalies, etc.
4. We look for the columns that contribute most to the model, both in
training time and goodness (score).
5. Find the best way to divide data, not always a method
It will work for all cases.
6. By creating pipelines of all steps, data loss is reduced,
improves accuracy and also maintains a modular structure and
flexible for cross validation execution and
later for the analysis of hyperparameters.
7. Choose the appropriate metrics, roc_auc, accuracy, confusion matrix, etc.
8. For production applications, the above steps are performed
many times, following a cycle, so that it can be polished
All the details of the model, making it reliable, robust and efficient.
9. Not everything is governed by the model with the highest score, many
Sometimes an explained model is required, which may have less score
but you can explain step by step your exits, like the trees of
decision.
10. Finally, data science is an art, and as such, it depends a lot
of the creativity of the analyst to obtain a very fine raw material
for subsequent training. Cleaner and meaningful is the data,
Your model will be better.
