# Demo: version 7

# Data science and Data engineering workflow

The idea is to show how a Data science (DS) team can work with autodeploy. Later,
how a Data engineering (DE) team can gather this insight and metadata to deploy 
a workflow to start making predictions in production.

1. The DS team deploys its workflows to see if everything runs as expected (in order to get the final model).

2. The DE needs some steps of those workflows in order to put the model in production.

3. Autodeploy allows the teams to create a minimal topology that best fist their needs.

4. Autodeploy will check their workflows over time to raise alerts if something goes wrong.


News: 
- Support architecture with networks, volumes.
- Tracker for each workflow works independently.
- Persistence is done for each node.
- Docker compose 90% compatibility with all workflows.
- Dashboard web front-end (20%)

##  Problem: Clasify leaf specimens

This data set consists of a collection of features that measure shape and texture extracted from digital images of leaf specimens from a total of 40 different plant species. It was collected in February 2014 at the Faculty of Science, Universidade do Porto, Portugal.https://archive.ics.uci.edu/ml/datasets/leaf

## 1. Build workflows

Build each environment inside each workflow


```python
from autodeploy.setup import setup

# App folder
app_dir = '/home/guess/Desktop/autodeploy/examples/demo1/data_science/'

# Workflows
workflow1 = [
    {'name': 'gathering', 'file': 'gathering.py', 
            'dockerfile': 'Dockerfile_gathering'}, # Provides a dockerfile
    
    {'name': 'preprocessing', 'file': 'preprocessing.py', 
            'requirements': 'req_preprocessing.txt'}, # Convert requirements.txt to dockerfile

]
workflow2 = [
    {'name': 'modeling', 'file': 'modeling.py', 
            'requirements': 'req_modeling.txt'}, # Convert requirements.txt to dockerfile
    
    
]
workflows = [
    {'name': 'workflow1', 'workflow': workflow1, 'tracker': {'port': 8001}},
    {'name': 'workflow2', 'workflow': workflow2, 'tracker': {'port': 8002}}
   
]

workflow_datascience = setup.Setup(app_dir, workflows)

workflow_datascience.build_workflows()
workflow_datascience
```

```python
workflow_datascience.workflows
```

## 2. Start workflows

Start containers belonging to each workflow

```python
# workflower.run_env(name='Workflower', port=8001, tracker=True)
workflow_datascience.start_workflows()
workflow_datascience
```

```python
workflow_datascience.workflows
```

```python
workflow_datascience.stop_workflows()
```

## Save environments for each workflow

```python
from autodeploy import tools

tools.create_registry('autodeploy_registry') 
    
# TODO: remote_path, visible machine for all teams platform
```

## 3. Run workflows

```python
from autodeploy.run import run

# Read the platform
runner = run.Run(workflow_datascience)

# Run the workflow
runner.run_workflows()
```

# Data engineering team

1. Build or read the workflows (build or read images)
2. Start the worklfows (start containers)
3. Run the workflows (run the python files on the containers)

## 1. Build workflow for predicting

```python
from autodeploy.setup import setup

# App folder
app_dir = '/home/guess/Desktop/autodeploy/examples/demo1/data_eng/'

# Workflows
workflow3 = [
    {'name': 'gathering_eng', 'file': 'gathering.py', 
            'env': 'gathering'}, # Provides an image from repository
    
    {'name': 'preprocessing_eng', 'file': 'preprocessing.py', 
            'env': 'preprocessing'}
]

workflows_eng = [
    {'name': 'workflow3', 'workflow': workflow3, 'tracker': {'port': 8003}},
]

workflow_dataeng = setup.Setup(app_dir, workflows_eng)

workflow_dataeng.build_workflows()
workflow_dataeng
```

```python
workflow_dataeng.start_workflows()
workflow_dataeng
```

```python
# workflow_dataeng.stop_workflows()
```

## Build predictor

Read model from environment

```python
from autodeploy.deploy import deploy

# Read the platform
deployer_eng = deploy.Deploy(workflow_dataeng)

# Build predictor
deployer_eng.build_predictor()

# Run predictor
deployer_eng.run_predictor()

deployer_eng
```

```python
# deployer_eng.stop_predictor()
```

## 2. Run workflows to start predicting

Once we have the workflows (gathering and preprocessing) and the predictor.

We can start making predictions from new data.

```python
from autodeploy.run import run

# Read the platform
runner_eng = run.Run(workflow_dataeng)

# Run the workflow
runner_eng.run_workflows()
```

```python
runner_eng.workflows
```

## Commit Predictor env into the registry

```python
# runner_eng.save_env(registry_name='autodeploy_registry')
```

## 3. Make Predictions

```python
import pandas as pd
import os

from autodeploy.check import tools

path = deployer_eng.app_dir
df_input = pd.read_csv(os.path.join(path, 'workflow', 'X_train.csv'))
df_noise = tools.add_noise(df_input, sigma=0.11) #requests


# df_cleaned = cleaner.run_workflow(df_noise, workflow=workflower_eng, env_name='Workflower') # missed step.

predictions = deployer_eng.predict(df_noise, port=5001, to_save=True, path_pred=path)
# tracker.save_prediction(path)

predictions.head()
# print(f'Predictions: \n {predictions}')
```

# Autodeploy will run these steps once configured 

## Checker type 1

Will check the features over time

```python
from autodeploy.check import check

checker = check.Checker(port=5001)

X_old = pd.read_csv(os.path.join(app_dir, 'workflow', 'X_train.csv'))
X_new = predictions.loc[:, predictions.columns != 'pred']

cols = ['eccentricity', 'elongation', 'solidity']

checker.run_checker(X_old.sample(len(X_new)), X_new, 
                   cols=cols, verbose=True,)

```

## New Checker (alpha) type 2

Will check all the features as a whole


New Checker based on autoencoders:


> This model will be applied to both new data and predictions (e.g, the model predicts a high loan).

> This new checker support realtime predictions (before was batch).

> Takes into account all features at once (before we analyzed one by one).

> Non linear decision boundaries (more complex models). Before we analyzed parametric, linear comparison.

> Unsupervised model. Before was also unsupervised, but less flexible to seasonality.
